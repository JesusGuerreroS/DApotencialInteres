# Proyecto de Análisis de Datos: Caso de estudio Potencial de ínteres 
### Autor: Jesús Guerrero
### Fecha: 27/07/2023

# Introduction
En este caso de estudio, voy a emplear las tareas principales de un análista de datos junior para la compañia sello07. En orden de responder las preguntas claves para el negocio, voy a seguir los pasos del proceso de análisis de datos: *Preguntar, Preparar, Procesar, Analizar, Compartir, y Actuar.* 

El reto principal a encarar en este proyecto, para resolver las preguntas planteadas, es la falta de datos, ya que la empresa es una startup y esta empezando crear su infraestructura.

# Antecedentes
## Sello07
Una startup del municipio Ayacucho, Edo Táchira, Venezuela. Con el objetivo de brindar el servicio para el mantenimiento de los tanques de agua. Se encuentra en el proceso de la captación de nuevos clientes, para iniciar labores en lo referente al cuidado de los tanques de agua, y así ayudar en gran medida a cuidar la salud de sus dueños, dado a las implicaciones que conllevan dichas practicas; además, contribuir al bienestar de las familias, gracias al mantenimiento de los depositos donde habita el vital líquido.

## Escenario
Soy un analista de datos junior, socio colaborador de la startup sello07. El CEO de la empresa quiere tener un mapa preciso, del potencial real que existe en las personas, para recibir el servicio del mantenimiento en sus tanques; para así dirigir de manera más eficiente los recursos a campañas de marketing y captación de nuevos clientes. Por lo tanto, mi equipo quiere saber dicho potencial en las principales localidades del municipio, con estos descubrimientos se tomarán decisiones en lo referente al diseño de nuevas estrategias de marketing, en lo referente a la captación de nuevos clientes; también, se tendrá un mapa más claro, hacia donde dirigir los recursos de la startup; pero primero se han de aprobar las recomendaciones respaldadas en datos y sus descubrimientos.

# Preguntar
## Tarea empresarial
Descubrir el potencial de ínteres que existe en las principales localidades del municipio San Juan de Colón, Estado Táchira, Venezuela.

## Preguntas del Análisis
Cuatro preguntas que guiarán las futuras estrategías de marketing, y uso de los recursos de manera más eficiente.  

1. ¿Qué potencial real existe en las personas, para realizarle un mantenimiento a sus tanques de agua?

2. ¿Con qué frecuencia, las personas de las principales localidades del municipio, le efectúan un mantenimiento a sus tanques de agua?

3. ¿Como incrementar el potencial de ínteres?

4. ¿Como está clasificada la población de tanques en las principales localidades?

5. ¿Qué localidades presentan el mayor potencial de ínteres?

# Preparar

## Fuente de los datos
Vamos a realizar una encuesta para la recolección de los datos, con una toma de muestra basada en la población de la locación elegida. Para una proporción representativa de la población, con un estimado de 65%, apuntando a un margen de error del 5% y un nivel de confianza del 95%. Vamos a tomar las 7 localidades principales del muncipio Ayacuho, para la recolección de los datos.

Para proteger la PII de los usuarios, los datos serán anonimizados y agregados; se le mostrará al participante de manera explícita, que se harán con los datos, por cuanto tiempo serán almacenados, para que serán usados y como resguardaremos su información personal.
